### Extracting Share Price (extracción del precio de las acciones)


A share is the single smallest part of a company's stock  that you can buy, the prices of these shares fluctuate over time. Using the <code>history()</code> method we can get the share price of the stock over a certain period of time. Using the `period` parameter we can set how far back from the present to get data. The options for `period` are 1 day (1d), 5d, 1 month (1mo) , 3mo, 6mo, 1 year (1y), 2y, 5y, 10y, ytd, and max.

$ \ $

Una acción es la parte más pequeña de las acciones de una empresa que puede comprar, los precios de estas acciones fluctúan con el tiempo. Usando el método `history()` podemos obtener el precio de las acciones durante un cierto período de tiempo. Usando el parámetro `period`, podemos establecer qué tan atrás del presente se obtienen los datos. Las opciones para `period` son 1 día (1 día), 5 días, 1 mes (1 mes), 3 meses, 6 meses, 1 año (1 año), 2 años, 5 años, 10 años, ytd y máx.


```python
apple_share_price_data = apple.history(period="max")
apple_share_price_data
```

The format that the data is returned in is a Pandas DataFrame. With the `Date` as the index the share `Open`, `High`, `Low`, `Close`, `Volume`, and `Stock Splits` are given for each day.

$ \ $

El formato en el que se devuelven los datos es Pandas DataFrame. Con la `Fecha` como índice, la acción `Apertura (abierto)`, `Alto`, `Bajo`, `Cierre`, `Volumen` y `Divisiones de acciones` se dan para cada día.

-----



A continuacion mostramos los primeros 5 datos.

```python
apple_share_price_data.head()
```

----

We can reset the index of the DataFrame with the `reset_index` function. We also set the `inplace` parameter to `True` so the change takes place to the DataFrame itself.

$ \ $

Podemos restablecer el índice del DataFrame con la función `reset_index`. También establecemos el parámetro `inplace` en `True` para que el cambio se realice en el DataFrame mismo.

```python
apple_share_price_data.reset_index(inplace=True)
apple_share_price_data.head()
```

-----

We can plot the `Open` price against the `Date`:

$ \ $

Podemos trazar el precio de `apertura` contra la `fecha`:


```python
apple_share_price_data.plot(x="Date", y="Open")
```

## Extracting Dividends


Dividends are the distribution of a companys profits to shareholders. In this case they are defined as an amount of money returned per share an investor owns. Using the method `dividends` we can get a dataframe of the data. The period of the data is given by the period defined in the 'history\` function.

$ \ $

Los dividendos son la distribución de las utilidades de una empresa a los accionistas. En este caso, se definen como una cantidad de dinero devuelta por acción que posee un inversor. Usando el metodo `dividendos`  podemos obtener un marco de datos de los datos. El período de los datos viene dado por el período definido en la función 'historial'.


```python
help(yf.Ticker.dividends) 
```

```python
apple.dividends 
```

We can plot the dividends overtime:


```python
apple.dividends.plot()
```
